```python
# importing the data from csv file to dataframe
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
amazon = pd.read_csv('/Users/inderpalsura/Desktop/Lighthouse/capstone_project/data/clean_data.csv', low_memory=False)
amazon.head()
```

```python
# The trend in the distribution of ratings suggests that the products tend to receive positive ratings, as indicated by the concentration of ratings in the 4.0 to 4.5 range. This is an important trend to consider, especially if you are conducting sentiment analysis or assessing customer satisfaction.
plt.figure(figsize=(8, 5))
sns.histplot(data=amazon, x='reviews.rating', bins=5, kde=True)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=amazon, x='reviews.rating', y='brand')
plt.xlabel('Rating')
plt.ylabel('Brand')
plt.title('Box Plot of Ratings by Brand')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=amazon, x='categories', order=amazon['categories'].value_counts().index[:10])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Top 10 Categories by Count')
plt.xticks(rotation=90)
plt.show()
#The plot identify which product categories are the most prevalent in dataset. For example, it appears that "Electronics" and "Clothing, Shoes & Jewelry" are among the most common categories.
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=amazon, x='reviews.date', y='reviews.rating')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.title('Rating Over Time')
plt.xticks(rotation=45)
plt.show()
# This type of plot is valuable for understanding how ratings evolve and for identifying periods of change or stability in customer sentiment
```

```python
sns.pairplot(amazon[['reviews.rating', 'reviews.numHelpful', 'reviews.doRecommend']])
plt.show()
```

```python
# Convert 'True' and 'False' to 1 and 0
amazon['reviews.doRecommend'] = amazon['reviews.doRecommend'].map({'True': 1, 'False': 0})

correlation_matrix = amazon[['reviews.rating', 'reviews.numHelpful', 'reviews.doRecommend']].corr()
print(correlation_matrix)
```

```python
# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# HeatMap Analysis of 'reviews.rating', 'reviews.doRecommend', 'reviews.numHelpful'

'reviews.rating' and 'reviews.doRecommend': The heatmap shows a strong positive correlation between 'reviews.rating' and 'reviews.doRecommend' (indicated by the warm color). This suggests that higher ratings are positively associated with a higher likelihood of recommending a product. This finding is consistent with what you'd expect: customers who rate a product highly are more likely to recommend it.

'reviews.rating' and 'reviews.numHelpful': The heatmap shows a weak positive correlation between 'reviews.rating' and 'reviews.numHelpful' (indicated by the relatively cooler color). This indicates a slight positive relationship, suggesting that products with higher ratings may tend to receive a slightly higher number of helpful votes. However, this correlation is weaker compared to the 'reviews.rating' and 'reviews.doRecommend' relationship.

'reviews.numHelpful' and 'reviews.doRecommend': The heatmap shows a weak positive correlation between 'reviews.numHelpful' and 'reviews.doRecommend.' This suggests that reviews with a higher number of helpful votes are somewhat more likely to be recommended. However, this correlation is weaker than the correlation between 'reviews.rating' and 'reviews.doRecommend.'

```python
# Find the maximum review rating
max_rating = amazon['reviews.rating'].max()

# Find the user(s) with the maximum rating
max_rating_users = amazon[amazon['reviews.rating'] == max_rating]['reviews.username']

# If there are multiple users with the maximum rating, you can list them all
if len(max_rating_users) > 1:
    print(f"Users with the maximum rating: {len(max_rating_users)}")
else:
    print(f"The user with the maximum rating is {max_rating_users.iloc[0]} with a rating of {max_rating}.")
```

```python
# Define the rating threshold (e.g., 4 for positive ratings)
rating_threshold = 4

# Create a new column 'positive_rating' with binary values (1 for positive, 0 for not positive)
amazon['positive_rating'] = (amazon['reviews.rating'] >= rating_threshold).astype(int)

# Create a new column 'negative_rating' with binary values (1 for negative, 0 for not negative)
amazon['negative_rating'] = (amazon['reviews.rating'] < rating_threshold).astype(int)
```

```python
# Create a countplot for positive and negative ratings
plt.figure(figsize=(8, 6))
sns.countplot(data=amazon, x='positive_rating', palette="Set1")

# Customize the plot
plt.xlabel('Rating Type')
plt.ylabel('Count')
plt.title('Count of Positive and Negative Ratings')
plt.xticks([0, 1], ['Negative', 'Positive'])

# Display the plot
plt.show()
```

```python
# Create a countplot for recommended products and did purchase
plt.figure(figsize=(8, 6))
sns.countplot(data=amazon, x='reviews.doRecommend', hue='reviews.didPurchase', palette="Set1")

# Customize the plot
plt.xlabel('Recommended Product')
plt.ylabel('Count')
plt.title('Recommended Product vs. Did Purchase')
plt.xticks([0, 1], ['No', 'Yes'])

# Add a legend
plt.legend(title='Did Purchase', labels=['No', 'Yes'])

# Display the plot
plt.show()
```

# Key Observations:

For Recommended Products (labeled as "Yes"):
The majority of customers appear to have made a purchase when the product is recommended (right segment is taller).
There are still some customers who did not make a purchase, even when the product is recommended.

For Not Recommended Products (labeled as "No"):
The number of customers who did not make a purchase for non-recommended products is higher compared to recommended products.
Fewer customers made a purchase for non-recommended products (left segment is taller).
